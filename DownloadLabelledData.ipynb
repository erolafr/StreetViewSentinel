# Roboflow labelled data analysis
The aim of this notebook is to demonstrate the process of downloading a dataset from Roboflow, processing its COCO annotations to extract image metadata including class-specific object counts the different classes, splitting the dataset into separate dataframes based on the presence of these objects, and visualizing the frequency distribution of the detected objects. The resulting dataframes are also made available for download.

Download a version of the dataset in Roboflow

```python
!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="oe8dI6KkneKZ8zEcoztT")
project = rf.workspace("invasivedemography").project("opuntiagsv_test")
version = project.version(5)
dataset = version.download("coco")
```

```python
import json
import os
import pandas as pd

# Path to the downloaded dataset
dataset_dir = dataset.location

# Path to annotation JSONs (Roboflow gives 3: train, valid, test)
ann_dir = dataset_dir

all_metadata = []

# Loop over each split (train/valid/test)
for split in ["train", "valid", "test"]:
    ann_file = os.path.join(ann_dir, split, "_annotations.coco.json")
    if not os.path.exists(ann_file):
        continue

    with open(ann_file, "r") as f:
        coco = json.load(f)

    # Make dict for object counts per image
    counts = {img['file_name']: 0 for img in coco['images']}

    # Count annotations
    for ann in coco['annotations']:
        image_id = ann['image_id']
        img_name = next(img['file_name'] for img in coco['images'] if img['id'] == image_id)
        counts[img_name] += 1

    # Collect metadata (filename, height, width, id, count, etc.)
    for img in coco['images']:
        # Extract original filename by removing the Roboflow suffix
        original_file_name = img["file_name"].split(".rf.")[0]

        all_metadata.append({
            "split": split,
            "image_id": img["id"],
            "file_name": original_file_name, # Use the original file name
            "width": img["width"],
            "height": img["height"],
            "object_count": counts[img["file_name"]]
        })

# Convert to DataFrame
df = pd.DataFrame(all_metadata)

# Save as CSV
df.to_csv("all_images_metadata.csv", index=False)

print("Saved metadata for all images as all_images_metadata.csv")
display(df.head())
```

## 2. Process Annotations and Extract Metadata

Process the downloaded COCO annotation files to extract image metadata and count the occurrences of 'agave' and 'opuntia' objects in each image.

```python
import json
import os
import pandas as pd

# Path to the downloaded dataset
dataset_dir = dataset.location

# Path to annotation JSONs (Roboflow gives 3: train, valid, test)
ann_dir = dataset_dir

all_metadata = []

# Loop over each split (train/valid/test)
for split in ["train", "valid", "test"]:
    ann_file = os.path.join(ann_dir, split, "_annotations.coco.json")
    if not os.path.exists(ann_file):
        continue

    with open(ann_file, "r") as f:
        coco = json.load(f)

    # Create a mapping from category ID to category name
    category_map = {cat['id']: cat['name'] for cat in coco['categories']}

    # Make dict for object counts per image for specific classes
    counts = {}
    for img in coco['images']:
        counts[img['file_name']] = {'agave': 0, 'opuntia': 0}


    # Count annotations for specific classes
    for ann in coco['annotations']:
        image_id = ann['image_id']
        category_id = ann['category_id']
        category_name = category_map.get(category_id)

        if category_name in ['agave', 'opuntia']:
            img_name = next(img['file_name'] for img in coco['images'] if img['id'] == image_id)
            counts[img_name][category_name] += 1


    # Collect metadata (filename, height, width, id, agave_count, opuntia_count, etc.)
    for img in coco['images']:
        # Extract original filename by removing the Roboflow suffix
        original_file_name = img["file_name"].split(".rf.")[0]

        all_metadata.append({
            "split": split,
            "image_id": img["id"],
            "file_name": original_file_name, # Use the original file name
            "width": img["width"],
            "height": img["height"],
            "agave_count": counts[img["file_name"]]['agave'],
            "opuntia_count": counts[img["file_name"]]['opuntia']
        })

# Convert to DataFrame
df = pd.DataFrame(all_metadata)

print("Metadata extracted and processed.")
display(df.head())
```

## 3. Create Separate Dataframes for Agave and Opuntia

Filter the main dataframe to create two new dataframes: one for images containing 'agave' and another for images containing 'opuntia'.

```python
# Create a new DataFrame for images with at least one 'agave' object
df_agave = df[df['agave_count'] > 0].copy()

# Create a new DataFrame for images with at least one 'opuntia' object
df_opuntia = df[df['opuntia_count'] > 0].copy()

# Display the first 5 rows of both dataframes
print("DataFrame for images with Agave:")
display(df_agave.head())

print("\nDataFrame for images with Opuntia:")
display(df_opuntia.head())
```

## 5. Visualize Object Frequency

Plot histograms to show the frequency of detected 'agave' and 'opuntia' objects in their respective datasets.

```python
import matplotlib.pyplot as plt

# Plot histogram for agave object counts
plt.figure(figsize=(10, 6))
plt.hist(df_agave['agave_count'], align='left', rwidth=0.8)
plt.xlabel('Number of Agave Objects')
plt.ylabel('Frequency')
plt.title('Frequency of Detected Agave Objects per Image')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Plot histogram for opuntia object counts
plt.figure(figsize=(10, 6))
plt.hist(df_opuntia['opuntia_count'], align='left', rwidth=0.8, color='orange')
plt.xlabel('Number of Opuntia Objects')
plt.ylabel('Frequency')
plt.title('Frequency of Detected Opuntia Objects per Image')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot overlapping histograms for agave and opuntia object counts
plt.figure(figsize=(10, 6))
plt.hist(df_opuntia['opuntia_count'], bins=20, alpha=0.7, label='Opuntia', color='orange')
plt.hist(df_agave['agave_count'], bins=20, alpha=0.7, label='Agave', color='blue')
plt.xlabel('Number of Objects')
plt.ylabel('Frequency')
plt.title('Frequency of Detected Agave and Opuntia Objects per Image (Overlap)')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```
